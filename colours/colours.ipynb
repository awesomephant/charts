```python
import pandas as pd
import numpy as np
from colorspace import sequential_hcl, diverging_hcl, qualitative_hcl, hclplot, specplot, rainbow_hcl, palette
from matplotlib import pyplot as plt

#https://retostauffer.github.io/python-colorspace/
```

```python
# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(7,6))
# hclplot(sequential_hcl()(17), ax=ax)

shades = {
    "violet": [
        "#E0D3F8",
        "#CDBAF0",
        "#BBA1E8",
        "#A889E1",
        "#9570D9",
        "#8257D1",
        "#6F3DC9",
        "#5820BF",
        "#441993",
        "#301268",
        "#1D0B40"
		]
}

pal = palette(shades["violet"])

# hclplot(pal.colors(), ax = ax1)
specplot(pal.colors())
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (10, 3.5))
specplot(sequential_hcl()(7),  ax = ax1)
hclplot(diverging_hcl()(7),   ax = ax2)
hclplot(qualitative_hcl()(7), ax = ax3)
plt.show()
```
